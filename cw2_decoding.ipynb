**Note on generation functions**

Generator functions are an advanced feature of Python that makes it easy to iterate over complicated datasets. The general syntax is just a standard function that uses the keyword ``yield`` instead of ``return`` to return data, which allows it to return data multiple times. You can iterate over the values returned by a generator function instead of just calling it. Here's an example:

````python
def gimme_some_multiples(n):
    yield n*1
    yield n*2
    yield n*3

for x in gimme_some_multiples(3):
    print(x)
````

````python
def gimme_some_more_multiples(n, how_many):
    for i in range(how_many):
        yield n*(i+1)

for x in gimme_some_more_multiples(5, 4):
    print(x)
````

````python

def bin_spike_times(spike_times, start_time, end_time, dt):
    """
    Bins spike times into a time window defined by [start_time, end_time]
    with a time step of dt.

    Args:
        spike_times (np.array): Array of spike times.
        start_time (float): The start time of the window.
        end_time (float): The end time of the window.
        dt (float): The time step to bin spikes at.

    Returns:
        np.array: A binary array indicating spike times (0 or 1).
    """
    num_bins = int((end_time - start_time) / dt)
    num_neurons = len(spike_times)
    binned_spikes = np.zeros((num_neurons, num_bins), dtype=int)

    # Bin spikes
    for neuron in range(num_neurons):
        for spike in spike_times[neuron]:
            if (start_time <= spike < end_time):
                bin_idx = int((spike - start_time) / dt)
                binned_spikes[neuron, bin_idx] = 1

    return binned_spikes

````

## Surrogate gradient descent spike function

Below is the code for the surrogate gradient descent function from lectures. You can use it as is, although note that there is a hyperparameter (scale) that you can experiment with if you choose.

````python
class SurrogateHeaviside(torch.autograd.Function):
    """
    Here we implement our spiking nonlinearity which also implements
    the surrogate gradient. By subclassing torch.autograd.Function,
    we will be able to use all of PyTorch's autograd functionality.
    Here we use the normalized negative part of a fast sigmoid
    as this was done in Zenke & Ganguli (2018).
    """

    scale = 100.0 # controls steepness of surrogate gradient

    @staticmethod
    def forward(ctx, input):
        """
        In the forward pass we compute a step function of the input Tensor
        and return it. ctx is a context object that we use to stash information which
        we need to later backpropagate our error signals. To achieve this we use the
        ctx.save_for_backward method.
        """
        ctx.save_for_backward(input)
        out = torch.zeros_like(input)
        out[input > 0] = 1.0
        return out

    @staticmethod
    def backward(ctx, grad_output):
        """
        In the backward pass we receive a Tensor we need to compute the
        surrogate gradient of the loss with respect to the input.
        Here we use the normalized negative part of a fast sigmoid
        as this was done in Zenke & Ganguli (2018).
        """
        input, = ctx.saved_tensors
        grad_input = grad_output.clone()
        grad = grad_input/(SurrogateHeaviside.scale*torch.abs(input)+1.0)**2
        return grad

# here we overwrite our naive spike function by the "SurrogateHeaviside" nonlinearity which implements a surrogate gradient
surrogate_heaviside  = SurrogateHeaviside.apply
````

````python
class SNNLayer(nn.Module):
  def __init__(self, n_in, n_out,  tau=50e-3, dt=1e-3, spiking=True , plot = False):

    """
    Initializes a single LIF layer.
    Args:
    n_in (int): Number of input neurons.
    n_out (int): Number of output neurons.
    tau (float): Membrane time constant in seconds.
    dt (float): Simulation time step in seconds.
    spiking (bool): Whether the output is spiking (True) or membrane potential (False).
    weight_scale (float): Scale for random weight initialization.
    """
    super(SNNLayer, self).__init__()
    self.n_in = n_in
    self.n_out = n_out
    self.spiking = spiking
    # Store weights as a trainable parameter
    self.tau = tau
    self.dt = dt
    # Having the weights as randoms instead of 1's allows us to see different velocities for an untrained model!
    self.w = nn.Parameter(torch.randn((n_in, n_out), device=DEVICE) * np.sqrt(1.0 / n_in))
    self.w.to(DEVICE)
    self.alpha = np.exp(-dt/tau)
    # Threshold and reset potential
    self.v_threshold = 1.0  # Spike threshold
    self.v_reset : float = 0.0      # Reset potential


  def forward(self, x):
    """
    Forward pass through the SNN layer.

    Args:
    x (torch.Tensor): Input spikes of shape (batch_size, n_in, num_time_points).

    Returns:
    y (torch.Tensor): Output spikes or membrane potentials of shape (batch_size, n_out, num_time_points).
    """
    batch_size, num_neurons, num_time_points = x.shape
    membrane_potential = torch.zeros((batch_size,self.n_out),device=DEVICE)
    y = torch.zeros((batch_size,self.n_out,num_time_points),device=DEVICE)

    for t in range(num_time_points):
      weighted_input_spikes = torch.matmul(x[:,:,t],self.w)
      membrane_potential= self.alpha*membrane_potential +  weighted_input_spikes
      spikes = surrogate_heaviside(membrane_potential - self.v_threshold)
      if self.spiking:
        y[:,:,t] = spikes
        membrane_potential = membrane_potential * (1 - spikes)
      else:
        y[:,:,t] = membrane_potential

    return y
````

````python

# Define the SNN with one hidden and one output layer
class SpikingNN(nn.Module):
    def __init__(self, n_input , n_hidden , n_output , tau_spike = 50 * ms , tau_non_spike = 500*ms,plot = False, skip_hidden = False):
        super(SpikingNN, self).__init__()
        self.skip_hidden = skip_hidden
        if not skip_hidden:
            self.hidden_layer = SNNLayer(n_input,n_hidden,spiking=True,tau = tau_spike)
            self.output_layer = SNNLayer(n_hidden,n_output,spiking=False , tau=tau_non_spike)
        else:
            self.output_layer = SNNLayer(n_input,n_output,spiking=False , tau=tau_non_spike)
        self.plt_hidden_layer = plot

    def forward(self,x):
        if not self.skip_hidden:
            hidden_layer_output = self.hidden_layer(x)
            output = self.output_layer(hidden_layer_output)
        else:
            output = self.output_layer(x)

        if self.plt_hidden_layer:
            # Plot 1 random batch
            batch_size , n_out , num_time_points = hidden_layer_output.shape
            batch_index = np.random.randint(batch_size)
            spikes = hidden_layer_output[batch_index].cpu().numpy()
            firing_rates = []
            for spike in spikes:
                firing_rates.append(spike.sum()/(num_time_points*1e-3))
            plt.figure(figsize=(10, 5))
            plt.xlabel('Firing Rates (Hz)')
            plt.ylabel('Frequency')
            plt.title('Firing Rate Distribution')
            plt.hist(firing_rates, bins=10, color='skyblue', edgecolor='black')
            print(f' Mean firing rate for hidden layer is {(spikes==1).sum()/(n_out*num_time_points*1e-3):.1f} Hz')
            self.plt_hidden_layer = False
        return output

# Function to compute loss
def evaluate_model(network,range, loss_fn=nn.MSELoss() , plot = False):
    with torch.no_grad():
        total_loss = 0
        total_null_loss = 0
        count = 0
        with torch.no_grad():
            for x, y in (batched_data(range_to_use=range, dt=1e-3, length=1, batch_size=64)):
                prediction = network(x)
                total_loss += loss_fn(prediction,y).item()
                total_null_loss += loss_fn(torch.zeros_like(prediction),y).item()
                count+=1
                if count ==1 and plot :
                    plt.figure(figsize=(15, 10))
                    x_velocities = prediction[0][0].cpu().numpy()
                    y_velocities = prediction[0][1].cpu().numpy()
                    plt.plot(x_velocities,label="X velocity",color="blue")
                    plt.plot(y_velocities,label="Y velocity",color="orange")
                    plt.xlabel('Time (s)')
                    plt.ylabel('velocity')
                    plt.legend(loc='best')
                    plt.title("all x- and y- components of the velocities")
                    plt.tight_layout()

        mean_loss = total_loss / count
        mean_null_loss = total_null_loss / count
        return mean_loss, mean_null_loss

# Function to plot internal outputs and firing rates
loss_fn=nn.MSELoss()
# Initialize the network
network = SpikingNN(n_input=52, n_hidden=100, n_output=2,plot=True)

# Evaluate the trained network
mean_loss, mean_null_loss = evaluate_model(network,testing_range,plot = True)
print(f"Mean Loss (untrained): {mean_loss:.4f}")
print(f"Mean Null Loss: {mean_null_loss:.4f}")
````

````python
def decoding_plot(decoding_range, dt_decoding=0.2, decoding_length=15, length=1, dt=1e-3, figdims=(4, 2)):
 # Defining time intervals for decoding windows
    w_intervals = np.arange(decoding_range[0] + length, decoding_range[0] + length + decoding_length, dt_decoding)
    batch_size_single = len(w_intervals) - 1  # Number of overlapping segments in each plot
    num_time_points = int(np.round(length / dt))  # Total time points per segment
    nfx, nfy = figdims  # Unpack subplot dimensions (number of rows and columns)
    nf = nfx * nfy  # Total number of subplots
    batch_size = nf * batch_size_single  # Total number of segments across all subplots

    with torch.no_grad():   # Disable gradient calculation for evaluation
        # Initialising tensors to store spike data and target velocities
        x = torch.zeros((batch_size, 52, num_time_points))  # Assuming 52 neurons
        y = torch.zeros((batch_size, 2))  # For storing x and y velocity targets
        T = []  # Time points for plotting

        # Interpolation functions for target velocities
        f_vel_x = interp1d(vel_times, vel_whitened[:, 0], bounds_error=False, fill_value="extrapolate")
        f_vel_y = interp1d(vel_times, vel_whitened[:, 1], bounds_error=False, fill_value="extrapolate")

        # Populate the input (spike data) and target (velocity data) tensors
        for b in range(batch_size):
            w_start = decoding_range[0] + dt_decoding * b  # Start time of the current window
            w_end = w_start + length  # End time of the current window
            T.append(w_end)  # Save the end time for plotting purposes

            # Looping through each neuron to bin spikes within the current window
            for n in range(len(spike_times)):
                spikes = spike_times[n]  # Spike times for the current neuron
                # Find spikes within the current window and convert to time indices
                indices = ((spikes[(spikes >= w_start) & (spikes < w_end)] - w_start) / dt).astype(int)
                x[b, n, indices] = 1.0  # Set spike times in the binary matrix

            # Interpolate the target velocities at the end of the window
            y[b, 0] = torch.tensor(f_vel_x(w_end))  # X velocity target
            y[b, 1] = torch.tensor(f_vel_y(w_end))  # Y velocity target

        T = np.array(T)  # Convert list of time points to a numpy array
        x, y = x.to(DEVICE), y.to(DEVICE)

        # Generate predictions using the neural network
        y_out = single_layer_network(x)[:, :, -1]  # Extract the final output time point for each segment
        mean_mse = torch.nn.functional.mse_loss(y, y_out).item()  # Calculate mean squared error (MSE)

        # Plotting
        fig, axs = plt.subplots(nfy, nfx, figsize=(nfx * 3, nfy * 2))
        axs = axs.ravel()  # Flatten the 2D grid of subplots for easier indexing

        # Loop through each subplot to plot the true and predicted velocities
        for nf_i in range(nf):
            start_idx, end_idx = nf_i *batch_size_single, (nf_i + 1) * batch_size_single   # Index range for the current subplot
            ax = axs[nf_i]  # Select the current subplot

            # Plot the target (true) and predicted X velocities
            ax.plot(
                T[start_idx:end_idx],
                y[start_idx:end_idx, 0].cpu(),
                label="True Velocity X",  # True X velocity
                linestyle=":"
            )
            ax.plot(
                T[start_idx:end_idx],
                y_out[start_idx:end_idx, 0].cpu(),
                label="Predicted Velocity X",  # Predicted X velocity
                linestyle=":"
            )

            # Set axis labels and titles for clarity
            ax.set_ylim(-4, 4)  # Set fixed y-axis limits for all plots
            ax.set_title(f"Plot {nf_i + 1}")  # Title for each subplot
            if nf_i % nfx == 0:
                ax.set_ylabel("Velocity")
            if nf_i >= (nfy - 1) * nfx:
                ax.set_xlabel("Time (s)")
            if nf_i == 0:
                ax.legend(loc="best")   # Add legend only to the first subplot

        # Adding main title showing the mean MSE value
        fig.suptitle(f"Mean MSE: {mean_mse:.3f}")
        plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit main title
        plt.show()  #Display

decoding_plot((0,100))
````
